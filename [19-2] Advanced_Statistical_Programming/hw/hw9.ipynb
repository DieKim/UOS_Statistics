# Q2

```python
import statistics
import math

weight = [65, 78, 80, 140, 100, 75, 78]

mean_w = statistics.mean(weight)
median_w = statistics.median(weight)
mode_w = statistics.mode(weight)
variance_w = statistics.variance(weight)
standard_w = math.sqrt(variance_w)

print("Answer2 is:")
print(mean_w, median_w, mode_w, variance_w, standard_w, sep=', ')
```

# Q3

```python
from matplotlib import pyplot as plt
import random

def num(n): # x축: 표본 수
    num = []
    for i in range(n):
        num += [i]
    return num

n = num(1000)

def cumulative_mean(n): # y축: 누적 평균
    cumul_mean = []
    a = []
    for i in range(n):
        a += [random.random()]
        mean = sum(a)/(i+1)
        cumul_mean += [mean]
    return cumul_mean

mean = cumulative_mean(1000)

xs = range(len(n))
plt.plot(xs, mean, 'b-')
plt.xlabel('n')
plt.ylabel('cumulative mean')
plt.title("Answer3")
plt.show() 

# n이 커질수록 누적 평균은 0.5에 가까워진다.
```

```python
from numpy import linalg as LA
import numpy as np
import random

# 3*3 행렬 생성 코드
def make_matrix():
    x1 = random.randint(1, 100)
    x2 = random.randint(1, 100)
    x3 = random.randint(1, 100)
    x4 = random.randint(1, 100)
    x5 = random.randint(1, 100)
    x6 = random.randint(1, 100)
    x7 = random.randint(1, 100)
    x8 = random.randint(1, 100)
    x9 = random.randint(1, 100)
    mat = np.array([[x1, x2, x3], [x4, x5, x6], [x7, x8, x9]])
    return mat

X = make_matrix()
Y = make_matrix()

print("X is:\n", X)
print("Y is:\n", Y)

# 두 행렬의 곱
print('multiplying X and Y with dot():\n', np.dot(X, Y))

# 두 행렬의 원소들의 곱
print('multiplying X and Y with array():\n', X*Y)
```
